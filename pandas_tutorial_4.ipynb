# join

# inner join(commens)

```python
week1.merge(week2 , how="inner" ,on="Customer ID" , suffixes=["week_1" , "week_2"])
```

```python
#people who came in both weeks and order the same foods
week1.merge(week2 , how="inner" ,on=["Customer ID","Food ID"] , suffixes=["week_1" , "week_2"])
```

# outer join(combination of two dataframe whole of records are there)

```python
merge=week1.merge(week2 , how="outer" ,on="Customer ID" ,suffixes=["week1" ,"week2"] , indicator=True)
#merge=week1.merge(week2 , how="outer" ,on="Customer ID" ,suffixes=["week1" ,"week2"] )
merge.head()
```

```python
merge["_merge"].value_counts()
```

```python
mask=merge["_merge"]!="both"
merge[mask]
```
