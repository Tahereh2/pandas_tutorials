# pivot table()

```python
food=pd.read_csv("D:\\tala\\pandas\\foods.csv" )
food.head()
```

```python
food.info()
```

```python
food.nunique()
```

```python
food["Gender"]=food["Gender"].astype("category")
```

```python
food["City"]=food["City"].astype("category")
food["Frequency"]=food["Frequency"].astype("category")
food["Item"]=food["Item"].astype("category")
```

```python
food.info()
```

```python
food.pivot_table(values="Spend" , index="Gender" ,aggfunc="mean")
```

```python
food.pivot_table(values="Spend" , index="City" ,aggfunc="mean")
```

```python
food.pivot_table(values="Spend" , index=["City" ,"Gender" ,"Item"] ,aggfunc="mean")
```

```python
food.pivot_table(values="Spend" , index=["City","Gender"] , columns=["Item" ,"Frequency"] ,aggfunc="mean")
```

# aggregate

```python
agg=pd.read_csv("D:\\tala\\pandas\\fortune1000.csv")
agg.head()
```

```python
agg.info()
```

```python
agg.set_index("Rank" , inplace=True)
```

```python
agg.head()
```

```python
agg.nunique()
```

```python
agg.groupby("Sector")
```

```python
group_sectors=agg.groupby("Sector")
```

```python
type(group_sectors)
```

```python
#how many groups
len(group_sectors)
```

```python
#size  or number of each group
group_sectors.size().sort_values(ascending=False)
```

```python
agg["Sector"].value_counts()
```

```python
#the first/last element of each group
group_sectors.first()
#group_sectors.last()
```

```python
#each groups memember numbers
group_sectors.groups
```

```python
#all memebor of eachgroup
group_sectors.get_group("Transportation")
#agg[agg["Sector"]]=="Transportation"
```

# sum in group by

```python
group_sectors.sum()
#group_sectors.get_group("Aerospace & Defense")["Revenue"].sum()
```

```python
#group_sectors.mean()
group_sectors["Revenue"].mean()
```

```python
group_sectors["Profits"].max().sort_values(ascending=False)
#group_sectors[["Profits" , "Revenue"]].min()
```

```python
#group by multy parameter
sectors=agg.groupby(["Sector" , "Industry"])
```

```python
sectors.size()
```

```python
len(sectors)
```

```python
sectors.sum()
```

```python
sectors.agg({"Revenue":"sum" ,"Profits":"mean" ,"Employees": "max"})
```

```python
sectors.agg(["sum" , "mean" ,"max", "min"])
```

```python
sectors.agg({"Revenue":["sum" ,"mean"] , "Profits":["max", "min"] ,"Employees":["sum","min","max","mean"]})
```
